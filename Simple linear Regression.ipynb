# Simple linear Regression explaing Gradient Descent

```python
import matplotlib.pyplot as plt
%matplotlib notebook

plt.style.use("ggplot")
# importing  vizulaisation dependencies 
```

```python
import numpy as np

# importing numpy for numric computatiom 
```

```python
X = np.linspace(0, 10, 20) + np.random.uniform(-2, 2, 20)
y = np.linspace(0, 10, 20) + np.random.uniform(-2, 2, 20)
plt.plot(X, y, "*")
    
'''
    taking 20 lineraly spaced values between 0 to 10 
    and adding a random noise to them 
    you can vizualize the the point coordinates in the below plot 
'''
```

All regression does is it fits a simple line to the point
i.e, 

##  Y = m * X + c 

we have X and need m, c to fit a line

```python
# m = np.random.rand(1)
# c = np.random.rand(1)
m = c = 0
print(m, c)
# chosing the random m, c values
```

```python
yhat = m * X + c
plt.plot(X, y, '*')
plt.plot(X, yhat)
print(((y - (m * X + c) )**2).sum())
```

## Gradient descent
- to chose the perfect parameter we run gradient descent over the cost function

# Error = (y - yhat) ** 2


yhat is the predicted value

```python

learning_rate = 0.001

loss = list()

plt.plot(X, y, "*")

for i in range(200):
    """
    caliculating the darivative with respect to m and c 
    chosing learning rate as 0.001 running gradient descent for twenty iterations
    appling the learning rate changing the parameter gives the perfect fit
    
    """
    # computing the derivatives
    dm = (X.dot(y)).mean() - (m * ( X.dot(X) ).mean() + c * X.mean())
    dc = y.mean() - (m * X.mean() + c)

    # appling gradient descent with learning rate
    t1 = m - learning_rate * (-2/20) * (dm) 
    t2 = c - learning_rate * dc * (-2/20)

    # updating the parameters
    m = t1
    c = t2
    
    yhat = m * X + c
    
    error = ((y - (m * X + c) )**2).sum()
    loss.append(error)
    plt.plot(X, yhat, alpha= 0.3 )

    
print(m, c)
```
